```
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',options=chrome_options)
driver.set_window_size(1400,1000)
```

```
#Paste Link of search in Livemint 
driver.get("https://www.livemint.com/Search/Link/Keyword/shree-cement")
div = driver.find_element_by_id('mySearchView')
links=div.find_elements_by_tag_name('span')
while('2014' not in links[-1].text):
  driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
  time.sleep(2)
  bod=driver.find_element_by_tag_name('body')
  div = driver.find_element_by_id('mySearchView')
  bod.send_keys(Keys.CONTROL,Keys.END)
  bod.send_keys(Keys.CONTROL,Keys.HOME)
  links=div.find_elements_by_tag_name('span')
  print(links[-1].text)
```

```
ypr=div.find_elements_by_tag_name('a')
z=[x.get_attribute('href') for x in ypr]
all=list(dict.fromkeys(z))
all.remove(None)
for x in all:
  print(x)
```

```
# Write the Name of the text file
f=open("Shree.txt",'w')
for x in all:
  try:
    html=urllib.request.Request(x,headers={'User-Agent':'Mozilla/5.0'})
    page=urllib.request.urlopen(html).read()
    soups=BeautifulSoup(page,"html.parser")
  except urllib.error.HTTPError as e:
    print('Error code: ', e.code)
    continue
  except urllib.error.URLError as e:
    print('Error code: ', e.code)
    continue
  except:
    print("Reading Error")
    continue
  else:
    pass
  dates=soups.find('span',id=re.compile("tBox.*"))
  try:
    date_obj = datetime.strptime(dates.get_text()[9:-14], '%d %b %Y')
    print(date_obj.strftime('%Y-%m-%d'))
    if(date_obj.year==2020 or date_obj.year<=2017):
      continue
    f.write(date_obj.strftime('%Y-%m-%d'))
  except:
    print("Date Extraction Error")
  else:
    pass
  f.write("\n")
  soups=soups.find_all('p')
  for z in soups:
      prev=''
      try:
        links=z.get_text()
      except:
        print("Article Extraction Error")
        continue;
      else:
        pass
      if("Click here to read the Mint ePaper" in links):
        break
      if(prev!=links):
        f.write(links+"\n")
        # print(links)
      prev=links
  f.write('\n****************************************************************************************************\n')
  # print('\n****************************************************************************************************\n')
f.close()
```
