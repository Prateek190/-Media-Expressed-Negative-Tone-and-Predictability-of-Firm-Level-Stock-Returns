```
# print(str(int(a.strftime("%d"))))
f=dict() 
s=["Adani", "Asian Paints", "Axis Bank", "Bajaj Auto", "Bajaj Finance", "Bajaj Finserv", "Airtel", "Bharti Infratel", "BPCL", "Britannia", "Cipla", "Coal India", "Reddy", "Eicher", "GAIL", "Grasim", "HCL", "HDFC", "HDFC Bank", "Hero MotoCorp", "Hindalco", "Unilever", "ICICI", "IndusInd", "Infosys", "IOC", "ITC", "JSW", "Kotak", "Larsen", "Mahindra & Mahindra", "Maruti", "Nestle", "NTPC", "ONGC", "PowerGrid", "Reliance", "Shree Cement", "State Bank", "Sun Pharma", "TCS", "Tata Motors", "Tata Steel", "Tech Mahindra", "Titan", "UltraTech", "Phosphorus Limited", "Vedanta", "Wipro", "Zee"]
start=[2, 2, 2, 2, 1, 1, 4, 1, 2, 1, 1, 2, 2, 1, 1, 1, 1, 1, 3, 1, 1, 2, 2, 2, 3, 2, 2, 2, 1, 2, 3, 3, 1, 2, 1, 1, 7, 1, 6, 1, 3, 3, 2, 1, 1, 1, 1, 2, 2, 1]
end=[13, 13, 27, 12, 8, 6, 28, 7, 11, 6, 14, 14, 19, 15, 11, 11, 15, 1, 23, 13, 11, 13, 42, 31, 32, 17, 18, 25, 16, 22, 32, 27, 10, 15, 7, 9, 36, 7, 38, 17, 32, 31, 31, 12, 11, 14, 5, 18, 21, 14]
adderfile=open("/content/drive/My Drive/Business/links.txt",'r')
adder=adderfile.readlines()
```

```
for x in s[42:]:
  f[x]=open("/content/drive/My Drive/Business/"+x+".txt",'w')
for i in range(42,50):
  print(i+1,s[i])
  for j in range(start[i],end[i]+1):
    print(j,'/',end[i])
    url="https://www.thehindubusinessline.com/search/?order=DESC&page="+str(j)+"&q=the&sort=publishdate"+adder[i][:-1]
    html=urllib.request.Request(url,headers={'User-Agent':'Mozilla/5.0'})
    page=urllib.request.urlopen(html).read()
    soup=BeautifulSoup(page,"html.parser")
    sou=soup.find('div',role="main").find_all('article')[:-1]
    for soups in sou:
      st=soups.find_all('a',href=re.compile("https://www.thehindubusinessline.com/.*/article[0-9]+\.ece"))
      for x in st[::3]:
        if '/companies/announcements/' in x['href']:
          continue
        try:
          html=urllib.request.Request(x['href'],headers={'User-Agent':'Mozilla/5.0'})
          page=urllib.request.urlopen(html).read()
          so=BeautifulSoup(page,"html.parser")
          soups=so.find_all('p')
        except urllib.error.HTTPError as e:
          print('Error code: ', e.code)
          continue
        except urllib.error.URLError as e:
          print('Error code: ', e.code)
          continue
        else:
          pass
        links=""
        try:
          head=so.find('h1',class_='tp-title-inf')
          dates=so.find('span',class_='update-date')
          date_obj = datetime.strptime(dates.get_text().strip(), '%B %d, %Y')
          if(date_obj.year==2020 or date_obj.year==2017):
            continue
          f[s[i]].write(date_obj.strftime('%Y-%m-%d')+'\n')
          f[s[i]].write(head.get_text()+'\n')
          for z in soups[1:]:
            text=z.get_text()
            if("Sign in here" in text or "Also read:" in text):
                break
            links+=text+"\n"
          # f[s[i]].write(a.isoformat()+'\n')
        except AttributeError:
          print("Attribute Error")
          continue
        f[s[i]].write(links)
        f[s[i]].write("\n***********************************\n") 
for x in s[42:]:
    f[x].close()
```
