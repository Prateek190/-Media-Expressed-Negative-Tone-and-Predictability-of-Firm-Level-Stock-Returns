```
#Change month in (year,month,day)
a=datetime.date(2018,5,27)
#Find article number on year,month,1st day of month
start=43247
i=27
for x in s:
  f[x]=open("/content/drive/My Drive/2018/May/"+x+"27.txt",'w')
#while i<=(number of days in the month)
while(i<=31):
  print(i)
  yea=str(a.year)
  mon=str(int(a.month))
  html=urllib.request.Request("https://economictimes.indiatimes.com/archivelist/year-"+yea+",month-"+mon+",starttime-"+str(start)+".cms",headers={'User-Agent':'Mozilla/5.0'})
  page=urllib.request.urlopen(html).read()
  soup=BeautifulSoup(page,"html.parser")
  y=soup.find("section",id="pageContent")
  if(not y):
    continue
  num=len(y.find_all('a',href=re.compile('.*\.cms')))
  ans=1
  for x in y.find_all('a',href=re.compile('.*\.cms')):
    print(ans,'/',num)
    ans+=1
    try:
      html=urllib.request.Request("https://economictimes.indiatimes.com/"+x['href'],headers={'User-Agent':'Mozilla/5.0'})
      page=urllib.request.urlopen(html).read()
      soups=BeautifulSoup(page,"html.parser")
    except urllib.error.HTTPError as e:
      print('Error on date',a.isoformat())
      print('Error code: ', e.code)
      continue
    except urllib.error.URLError as e:
      print('Error on date',a.isoformat())
      continue
    else:
      pass
    soups=soups.find_all('div',class_="Normal")
    links=""
    for t in soups:
      links+=t.get_text()
    for op in s:
      for u in dic[op]:
        if u.lower() in links.lower():
          f[op].write(a.isoformat()+'\n')
          f[op].write(links.encode('utf-16', 'surrogatepass').decode('utf-16'))
          f[op].write("\n***********************************\n")
          # print(op)
          break
  start+=1
  a+=datetime.timedelta(days=1)
  i+=1
for x in s:
  f[x].close()
```
